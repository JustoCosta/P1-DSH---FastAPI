```python
import pandas as pd
```

```python
Laptime1 = pd.read_csv('Datasets\lap_times\lap_times_split_1.csv' , sep = ',' , header = None , decimal = '.')
Laptime1
```

```python
Laptime2 = pd.read_csv('Datasets\lap_times\lap_times_split_2.csv' , sep = ',' , header = None , decimal = '.')
Laptime2
```

```python
Laptime3 = pd.read_csv('Datasets\lap_times\lap_times_split_3.csv' , sep = ',' , header = None , decimal = '.')
Laptime3
```

```python
Laptime4 = pd.read_csv('Datasets\lap_times\lap_times_split_4.csv' , sep = ',' , header = None , decimal = '.')
Laptime4
```

```python
Laptime5 = pd.read_csv('Datasets\lap_times\lap_times_split_5.csv' , sep = ',' , header = None , decimal = '.')
Laptime5
```

Concatenar los Dataframes

```python
Laps = pd.concat([Laptime1 , Laptime2 , Laptime3 , Laptime4 , Laptime5])
Laps.index = range(Laps.shape[0])
Laps
```

```python
Laps.to_csv('Laps.csv' , sep = '|', encoding = 'utf-8' , index = False)
```
